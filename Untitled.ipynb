```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.hindustantimes.com/delhi-news/delhiwale-sweet-sorrows-in-the-times-of-corona/story-CmSC0eRA6OVcmXRRjENnDO.html"
request = requests.request("GET", url)
```

```python
print(request.text)
```

```python
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
print(soup.title.name)
```

```python
print(soup.title.parent.name)
```

```python
print(soup.p)
```

```python
print(soup.p['class'])
```

```python
print(soup.a)
```

```python
print(soup.find_all('a'))
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
